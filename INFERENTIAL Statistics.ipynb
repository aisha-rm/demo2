## Hypothesis Testing
- testing assumptions about sample data or population.
It is done by analyzing random sample of the data to decide whether to accept or reject the assumption.

 ### Types of hypothesis testing

    1. Null hypothesis- assumptions made on sample data purely at random. It is denoted by H0
    
    2. Alternate hypothesis- opposite of null hypothesis. assumptions aren't made at random. denoted by H1
    

**Steps of testing hypothesis**

Set significance level that determines the probability that the relationship between 2 or more variable is caused by more than chance or random occurence. denoted by alpha.


1. **State your research hypothesis** as null (H0) and alternate Hypothesis (H1)

eg
H0 - men are on average NOT taller than women
H1- men are on average taller than women

2. **Collect data in a way designed to test hypothesis** - i.e the data should be random to 
eg there should be equal proportion of men and women across
each variable that can influence result should be in equal proportion


3. **Perform an appropriate statistical test**
there are a variety, they are based on the comparison of how spread out the data is within a category versus how different the categories are from one another.

t- test ; used to compare means of 2 groups
p- value; is a number that describes the chances of finfing a particular set of observations if the null hypothesis were true. 

low p value- difference is likely not due to chance.
high p value- likely that the difference between the groups is due to chance.
p value- describes the chances of finding a particular set of observations if the null hypothesis were true.

4. **Decide if the null hypothesis is True or False**
if p value is less than significance level, null pypothesis is false
