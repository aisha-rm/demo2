Adding elements to a list
Append method- add one element to the end of the list
Insert methof- add element to a specific position
Loop method- add multiple elements 

```python
list = []
print("blank list" , list)
for i in range(1,10):
    list.append(i)
    
print(list)
```

```python
tuple1 = (2, 4, 5)
tuple2 = ("ana", "eesh")
tuple = tuple1 + tuple2
print(tuple)
tupple = (tuple1, tuple2)
print(tupple)
print(tuple[:-1])
print(tuple[::-1]) #double colon reversers the order
```
