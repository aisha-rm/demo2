VARIANCE/STANDARD DEVIATION

SD- measures the variation or dispersion of data points in data set.

```python
a = ([20, 30, 40, 50, 60])
np.std(a)
```

```python
np.mean(a)
```

```python
b = ([5, 20, 45, 60, 70])
print(np.mean(b))
print(np.std(b))
```

Variance= each sample minus the average, take square of each of the outcome, then find the mean of the numbers. this is variance. The square root of this number is the std deviation.

The higher the standard deviation
