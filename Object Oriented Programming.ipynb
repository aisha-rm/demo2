```python
class Cat:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    #adding method  
    def meow(self):
        print("Meow!")

    #functionality- info you want contained
    def catInfo(self):
        print(self.name + " is " + str(self.age) + " year(s) old")
        
my_cat = Cat("Ludacris", 1)        
        
my_cat.catInfo()
```
