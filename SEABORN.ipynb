```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris_data = sns.load_dataset("iris")
f = plt.figure(figsize = (6,4))
fig = sns.scatterplot(x = "sepal_length", y = "sepal_width", hue= "species", data = iris_data)
```
