```python
# notice the data above has no header, so the first row was printed on top as header
#notice data separated by "\t", so can add it to the argument when opening the csv file

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data-original"
webF2 = pd.read_csv(url, sep = "\t", header = None)  #header= None tells it there is no header, so doesnt move the first row as column
webF2
```

```python
webF2.head(15)
# notice column 0 is lots of colums cramped into one,so will need to be split
```

```python
#create copy just incase you run into issues, you dont need to start afresh
webF2copy = webF2.copy()
```

```python
#splitting Data
# cant use loc since the data is not named

webF2copy.iloc[:, 0]  #retrive column 0

#convert the column to a str
webF2copy.iloc[:, 0].str

#split the string using split method
webF2copy.iloc[:, 0].str.split()

#store the new list in to a new or the same variable
webF2copy = webF2copy.iloc[:, 0].str.split(expand=True)   # expand=true makes it print like a neat data frame, otherwiseprints cramped up
```

```python
webF2copy.head()
```

```python
# pd.concat adds 2 tables together normally row-wise. To add them column wise, you need to add axis into the argument 
```

```python
carName = webF2.iloc[:,1]  #retrieve car name column from main data
carName.head()
```

```python
cleanData = pd.concat([webF2copy, carName], axis = 1)  #y axis = 1. not putting the axis makes the concatenation row-wise by default. Axis 0 is for row.
cleanData.head()
```

cleanData.columns = headers

headers is a list of strings representing the name of the columns


```python
headers= ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model year", "origin", "car name"]
cleanData.columns = headers
cleanData.head()
```

```python
cleanData.info()

#note data type here are all object(i.e string) because we converterted the colum to string earlier, meanwhile some are actually floats and ints.
#non-null means its not empty 
```

1. mpg-continuous
2. cylinders - multi-valued discrete
3. displacement- continous
4. horsepower- continuous
5. weight- continuous
6. acceleration - continuous
7. model year - multi valued discrete
8. origin multi valued discrete
9. car name - unique for each instance

continous = floats, discrete = integers, object = string

Becuase we changed those columns to str when we were splitting above to clean data, we need to change them back to their normal format listed above

```python
#continous data being changed from object to floats
cleanData[["mpg", "displacement", "horsepower", "weight", "acceleration"]] = cleanData[["mpg", "displacement", "horsepower", "weight", "acceleration"]].astype('float')
```

```python
#the above error means some missing values when collecting the datawere filled as "NA". and program cannot convert this
#string to float. So we need to find the column these values and replace them with something valid

cleanData[cleanData.eq("NA").any(1)] #finding the culprits
```

above the horsepower and mpg columns have NA

so we have to replace it. usually  using the mean or most frequent number in the columns which can be found as follows.

```python
cleanData.describe()
```

most frequent is 22 in horsepower, the highest is 150.

```python
cleanData["horsepower"] = cleanData["horsepower"].replace("NA", "22") 
```

```python
#converting the int objects that are currently strings to their rightful int

cleanData[["cylinders", "model year", "origin"]] = cleanData[["cylinders", "model year", "origin"]].astype("float")   #need to turn to float first because there is value "8." in cylinders which will throw error as it wont convertto int directly
```

```python
cleanData[["cylinders", "model year", "origin"]] = cleanData[["cylinders", "model year", "origin"]].astype("int")
```
