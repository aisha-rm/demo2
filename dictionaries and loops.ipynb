**Control Statements**
continue- returns control to the beginning of the loop
break- brings control out of the loop

```python
for letter in "aisha":
    if letter == "s" or letter == "h":
        continue
    print(letter)
```

```python
for letter in "aisha":
    if letter == "i":
        break
    print(letter)
```
