#### Types of Data Used

1. Labelled - feautures (input variable) and outcomes (output or target variable) are labelled. It can be tasking as labels are made by humans

- Balanced - for labels with categorical data. balanced if the sample is equally spread across classes.
- Unbalanced- more data available in one class than other classes. This creates a bias model because there are more samples in on class that it learned with.

2. Unlabelled -features are labelled but outcomes are not, allowing the machine to learn what it can. There needs to be a large amt of data for this so that the machine can learn patterns

### SUPERVISED LEARNING
#### Types of Supervised Learning

1. Classification problem - used if outcomes are categorical. eg tribes, spam email classification etc
Goal is to correctly predict class of a new sample.
- Binary ; 2 classes of outcomes
- Multiclass; 3 or more classes
Evaluation metrics - AUC, F1 score, Average accuracy. 

2. Regression problem - predicts numerical values eg salary, loam, age etc
Evaluation metrics - MSE and RMSE. what is considered is how far apart the predicted value is from the actual value, bcos it is near impossible to predict the eaxt value.

### EVALUATION METRICS

1. #### Confusion Matrix - shows ratio of accurate predictions to wrong predictions in a binary classification

           Actual Values
           ==============
           Positive | Negative
               TP   | FP
               -------------
               FN   | TN
               
               

The y axis is predicted values


**Accuracy** - percentage of correctness in the models predcition. It doesnt account for biases in classification.

Accuracy = (TP + TN)/(TP+FP+TN+FN)  ie true values (pos and neg) divided by total sample.

**Precision** shows how many correctly predicted belong to the positive class. tells how reliable the model is.

P = TP/ (TP+FP)

**Recall** shows how many of the actual positive samples the model correctly predicted

R = TP/(TP+FN)

**F1 score** - combines precision and recall. For when you dont know which to go with. Gives a harmonic mean.

= 2/((1/recall)+1/(precision))

**AUC - ROC Curve** - show performance of model esp binary classification. AUC is area under curve. The higher the curve i.e closer to 1, the better.i.e it differentiates between 2 classes very well. If opposite, it means it is mistaking ine class for another.And if runs at 0, it means it found no pattern

If a model performs well on new samples then the model **generalises*. 
- Overfitting - when complex algorithms are applied to simple data sets, they learn the training perfevtly like cramming, but behave poorly on new samples
- Underfittign - simple algorithms used on co,plex data, they dont learn well, so they do poorly on new data. 

Accuracy of trained data and accuracy of test data are similar means the mdel is good Same for other evaluation metrics.
