**INTERQUARTILE RANGE**

PERCENTILES- 25, 50 AND 75TH



IR similar to SD. shows how spread out data is. often not as affected by outliers like SD is.

IQR = Q3 - Q1

Need to arrange the numbers in ascending order first.
Find median and isolate it.
Work with rest of the data i.e median of the numbers on each side

```python
data.sort()  #sorts in ascending order
```

```python
data
```

median is 60,
median of 1, 50, and 55 is 50 which is Q1 or 25th quartile,
Q3 is the median of numbers to the right i.e 86. it is the 75th percentile. Q2 is 50th percentile.
IQR = 86 - 50 = 36

```python
data3 =[45, 50, 55, 60, 74, 86, 90]
```

the above gives same IQR as first example even though not otliers. Sothe outliers in the first example didnt affect it much. That is its advantage unlike how the outliers affect mean and SD. However if,the outliers are large or many, then it may begin to affect IQR.

```python
#first quartile. 25 below is the 25th percentile
q1 = np.percentile(data, 25, interpolation = 'midpoint')
q1
```
