SERIES

```python
import numpy as np
import pandas as pd

array1 = np.array(["j", "f", "m", "a", "m", "j", "j", "a", "s", "o", "n", "d"])
series1 = pd.Series(array1)
series1
```

*NAMING THE SERIES*

```python
pd.Series(array1, name = "months")
```

To extract info from the data, can store into a new variable as below

```python
selectMonths = series1[3:7]
selectMonths
```

Customizing Index 

```python
pd.Series(array1, name = "months", index = ["a", "b", "c","d","e","f","g","h","i","j", "k","l"])
                                            
```

**OPERATIONS**

```python
data1 = [1,2,3]
data2 = [2,3,4]
numS1 = pd.Series(data1)
numS2 = pd.Series(data2)
numS1.add(numS2, fill_value = 0)
```

```python
numS1.mul(numS2, fill_value = 0)
```

```python
numS2.div(numS1, fill_value = 0)
```

```python
numS2.sub(numS1, fill_value = 0)
```

**DATA FRAMES**

```python
datalist = [["GA", 3], ["F", 2], ["UOE", 3]]
dataframe =  pd.DataFrame(datalist)
dataframe
```

```python
dataframe.columns = ["Course", "Grade Point"]
dataframe
```

```python
dataframe["Lecturers"] = ["Aisha", "Bishr", "Hadiza"]
dataframe
```

*Can create data frame from dicts too same like lists above*

```python
dataframe = pd.DataFrame(datalist, columns = ["course", "grade"])
dataframe
```

*INDEXING*

```python
dataframe["course"]
```

**Selecting rows in data frames**

iloc or loc

iloc is position based. using the default index.

loc is label based. using the names of rows and columns

For both, iloc or loc.[row_index, column_index]

```python
dataframe.index = ["1st", "2nd", "3rd"] #renaming index
dataframe
```

```python
dataframe.loc["2nd", "course"]
```

```python
dataframe.iloc[1, 0]
```

Extracting the first couple of rows

```python
dataframe.loc["1st":"2nd"]
```

```python
dataframe.iloc[0:2]
```

For 3 or more column data frame

```python
dataframe.iloc[0:2, 1:2]
```

[rows, columns]----------[:]  empty bracket returns the whole rows or columns

```python
dataframe.loc[:, "grade"]
```
