# SQL

Structured Query Language is a language used to query, manipulate data from database.

Database is collection of tremendous amount of related data or tables coming together to form an entity

Relatational database-

Relational key- 

Primary key - is a column or set of columns whose values uniquely identify a row in the table. A relational database is designed to enforce the uniqueness of primary keys by allowing only one row with a given key value in a table. It coud be an integer used as the unique ID

Foreign Key- values correspond to the values of the primary key in another table. In order to add a row witha given foreign key value, there must exist a row in the related table  with the same primary key value.

```python
#Using Tables

#How to get a column from database
SELECT column
FROM table

eg
SELECT continent
FROM countries

#2 or more columns
SELECT column1, column2, column3 etc
FROM tablename
```

```python
#how to get more specific details

SELECT column1, column2, column3 etc
FROM tablename
WHERE column1 = 'Value'

eg
SELECT continent, location, population
FROM countries
WHERE location = 'Afghanistan'
or
WHERE population < 300000000
or 
WHERE location != 'Afghanistan'

#getting all columns  uses astesk
SELECT *
FROM countries
#if condition needed
WHERE location = 'Afghanistan' AND continent = 'Asia' AND median age>25
# can also use OR as in booleans
```

```python
#Filtering and sorting data

SELECT DISTINCT location  #gives unique values as some locations may be replicated in the data set
FROM countries

SELECT location
FROM countries
WHERE NOT location = 'Afghanistan' # NOT is same as !=

ORDER BY continent  # reorders the continent data alphabetically
LIMIT 10   # Shows first 10 rows


SELECT *
FROM orders
ORDER BY account_id, total_amt_usd DESC # sorts it based on largest amounts per account/user
LIMIT 20

SELECT *
FROM orders
ORDER BY total_amt_usd DESC, account_id # sorts it based on largest amounts ever, then if 2 or more users spents same amt, it then sorts the account too
LIMIT 20

```

```python
#changing data

UPDATE countries
SET location = 'Denmark'
WHERE iso_code = 'ALB'


SELECT COUNT(iso_code)
FROM countries
#for conditions if needed
WHERE continent = 'Asia'


SELECT sum(population)
FROM  countries
WHERE continent = 'Asia'

SELECT avg(population)
FROM  countries
WHERE continent = 'Asia'
```

### SUBQUERY
a query inside a query is a subquery.

FROM countries

        FROM countries)

```python
#### joining columns

SELECT location, population, total_cases
FROM countries C                  # renames the table to C
JOIN cases ca                #renames the table to ca
ON countries.iso_code = cases.iso_code       # joins on the isocode that comes with countries and isocode in cases. This is what they have in common and are joined on.
GROUP BY location   # compresses multiple places with same name into one
ORDER BY population   # ascending order
ORDER BY population DESC # descending order

SELECT c.location, c.population, ca.total_cases, d.total_deaths
FROM countries c
JOIN cases ca
ON c.iso_code = ca.iso_code
JOIN deaths d
ON d.iso_code = c.iso_code #choose any of the already joined tables
```
